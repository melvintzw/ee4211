```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df_all = pd.read_csv('austin_weather.csv')
```

```python
display(df_all)
```

```python
df_all['Date'] = pd.to_datetime(df_all['Date'])
df_all = df_all.set_index(pd.to_datetime(df_all['Date']))
df_all = df_all.drop(columns='Date')
```

```python
display(df_all)
```

```python
df_period = df_all.loc['2015-10-1':'2016-3-31'] 
```

```python
display(df_period)
```

```python
df_plot = df_period['TempAvgF']
print(df_plot)
```

```python
plt.plot(df_plot)
```

```python
df_celsius = df_plot.apply(lambda x: (x - 32) * (5/9))
```

```python
plt.title('Average Temperature')
plt.plot(df_celsius)
```

```python
df_plot2 = df_period['HumidityAvgPercent']
```

```python
#plt.title('Average Humidity')
plt.figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')
plt.tight_layout()
plt.plot(df_plot2)
```

```python
df_plot3 = df_period['WindAvgMPH']
```

```python
plt.title('Average wind speed')
plt.plot(df_plot3)
```

```python
stack = pd.concat([df_plot,df_plot2],axis=1)
```

```python
display(stack)
```

```python
stack = pd.concat([stack,df_plot3],axis=1)
```

```python
display(stack)
```
